```python
TBills = pd.read_csv('/content/drive/MyDrive/ML_projects/QLearningBot/bill-rates-2002-2021.csv',parse_dates=['Date'],index_col='Date')
print(TBills.head())
```

```python
SNP500 = pd.read_csv('/content/drive/MyDrive/ML_projects/QLearningBot/sp500_stocks.csv',parse_dates=['Date'],index_col='Date')
print(SNP500.head())
```

```python
SNPandTBills = pd.merge(SNP500,TBills,left_index=True,right_index=True).sort_index()
print(SNPandTBills.head())
```

```python
SNPandTBills = SNPandTBills.dropna()
```

```python
SNPandTBills.shape
```

```python
SNPandTBills.head()
```

```python
SNPTBill = SNPandTBills.groupby(['Date']).agg({'Adj Close':'mean','4 WEEKS BANK DISCOUNT':'mean'})
```

```python
SNPTBill.head()
```

```python
SNPTBill['Stocks'] = SNPTBill['Adj Close'].rolling(window=2).apply(lambda x: x[1]/x[0] - 1)
SNPTBill['TBills'] = (SNPTBill['4 WEEKS BANK DISCOUNT']/100 + 1) **(1/4) - 1 
```

```python
SNPTBill.head()
```

```python
SNPTBill['risk_adjusted'] = SNPTBill['Stocks'] - SNPTBill['TBills']
```

```python
SNPTBill.head()
```

```python
SNPTBill = SNPTBill.dropna()
```

```python
SNPTBill.head()
```

```python
returns = SNPTBill.drop(['Adj Close',	'4 WEEKS BANK DISCOUNT'],axis=1)
```

```python
returns.head()
```
